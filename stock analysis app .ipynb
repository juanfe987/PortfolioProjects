```python
# Step 1: Install Required Libraries (run this once if not installed)
# !pip install yfinance pandas matplotlib seaborn

# Step 2: Import Required Libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 3: Load Stock Data
ticker = 'AMD'
stock = yf.Ticker(ticker)
data = stock.history(period="1y")

# Step 4: Get Financial Metrics
info = stock.info
pe_ratio = info.get('trailingPE')
forward_pe = info.get('forwardPE')
peg_ratio = info.get('pegRatio')
eps_growth = info.get('earningsQuarterlyGrowth') * 100 if info.get('earningsQuarterlyGrowth') else None
beta = info.get('beta')

# Scoring Functions
def score_pe(pe): return 3 if pe < 15 else 2 if pe < 25 else 1
def score_eps_growth(eps): return 3 if eps > 20 else 2 if eps > 5 else 1
def score_peg(peg): return 3 if peg < 1 else 2 if peg < 2 else 1

# Step 5: Create Score Table
metrics = {
    "P/E Ratio": [pe_ratio, score_pe(pe_ratio)],
    "Forward P/E": [forward_pe, score_pe(forward_pe)],
    "PEG Ratio": [peg_ratio, score_peg(peg_ratio)],
    "EPS Growth %": [eps_growth, score_eps_growth(eps_growth)],
    "Beta": [beta, "-"]
}
df_metrics = pd.DataFrame(metrics, index=["Value", "Score"]).T
print("=== Valuation & Growth Metrics ===")
print(df_metrics)

# Step 6: Final Verdict
score_total = df_metrics[df_metrics["Score"] != "-"]["Score"].sum()

def get_verdict(score):
    if score >= 9: return "✅ Buy"
    elif score >= 6: return "👀 Watch"
    else: return "❌ Avoid"

verdict = get_verdict(score_total)
print(f"\n📌 Final Score: {score_total} — Verdict: {verdict}")

# Step 7: Plot Price Chart
plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label='Close', color='blue')
plt.title(f'{ticker} Stock Price (1Y)')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid()
plt.show()

# Step 8: Moving Averages
data['50_MA'] = data['Close'].rolling(50).mean()
data['200_MA'] = data['Close'].rolling(200).mean()

plt.figure(figsize=(10, 5))
plt.plot(data['Close'], label='Close')
plt.plot(data['50_MA'], label='50-day MA', linestyle='--')
plt.plot(data['200_MA'], label='200-day MA', linestyle=':')
plt.title(f'{ticker} - Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend()
plt.grid()
plt.show()

# Step 9: RSI (Momentum Indicator)
def compute_RSI(data, window=14):
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=window).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=window).mean()
    RS = gain / loss
    RSI = 100 - (100 / (1 + RS))
    return RSI

data['RSI'] = compute_RSI(data)

plt.figure(figsize=(10, 4))
plt.plot(data['RSI'], label='RSI', color='purple')
plt.axhline(70, linestyle='--', color='red', label='Overbought')
plt.axhline(30, linestyle='--', color='green', label='Oversold')
plt.title(f"{ticker} - RSI (14-day)")
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.grid()
plt.show()
```
